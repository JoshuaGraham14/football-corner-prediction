```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#CONSTS
READ_FOLDER_PATH = '../'
WRITE_FOLDER_PATH = ''

data = pd.read_csv(READ_FOLDER_PATH+'totalCorner_odds.csv')

print(data.shape[0])
```

```python
#Select only numeric columns
data_filtered = data.drop(columns=["kaggle_id", "totalCorner_id"], errors="ignore")

#Create subplots for each column
for col in data_filtered:
    fig,axes = plt.subplots(1,2, figsize=(12, 5))

    #Histogram (left)
    axes[0].hist(data[col], bins=30,edgecolor='black',alpha=0.7)
    axes[0].set_title(f"Distribution of {col}")
    axes[0].set_xlabel(col)
    axes[0].set_ylabel("Frequency")
    axes[0].grid(axis="y", linestyle="--", alpha=0.7)

    #Boxplot (right)
    axes[1].boxplot(data[col].dropna(), vert=False,patch_artist=True)
    axes[1].set_title(f"Boxplot of {col}")
    axes[1].set_xlabel(col)
    axes[1].grid(axis="x", linestyle="--", alpha=0.7)

    plt.tight_layout()
    plt.show()
```

```python
max_value = data["odds_1_plus_corner"].max()
print(f"📈 Highest odds for 1+ corner: {max_value}")
```
